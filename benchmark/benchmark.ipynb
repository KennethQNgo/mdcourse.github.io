```python
import numpy as np

paths = ["../lammps-input/", "../molecular-simulation/"]

for path in paths:
    dump = path + "dump.lammpstrj"
    file = open(dump, "r")
    velocity = []
    norm_velocity = []
    for line in file:
        try:
            id, type, x, y, z, vx, vy, vz = np.float32(line.split())
            velocity.append([vx, vy, vz])
            norm_velocity.append(np.sqrt(vx**2+vy**2+vz**2))
        except:
            pass
```
